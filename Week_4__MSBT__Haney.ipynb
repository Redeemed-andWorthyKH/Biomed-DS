```python
#create a numeric list
import numpy as np

# Step 1: create a numeric list
numbers = [10, 3, 7, 1, 9, 5]

# Step 2: Sort the list in increasing order
sorted_numbers = sorted(numbers)

print("original list:" , numbers)
print("New sorted list (using sorted()):", sorted_numbers)

# Remove two values (example: remove 2 and 9)
numbers.remove(3)
numbers.remove(9)

# print

print("Reduced list:",numbers)
```
